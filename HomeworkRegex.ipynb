Задание 1.
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
import re
regex_number = r'([А-Я]{1}\d{3}(?<!000)[А-Я]{2})(\d{2,3})$'
def auto_number(regex_number):
    user_input = input('Введите транспортный номер ')
    result = re.search(regex_number, user_input)
    if re.match(regex_number, user_input):
        return(result.group(1), result.group(2))
    else:
        print('Номер не валиден.')
print(auto_number(regex_number))
```

Задание 3. **
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
phrase = 'Я учусь учусь работать в питоне питоне.'
def repeat(phrase):
    new_phrase = re.sub(r'(\b\w+\b)\W+\1', r'\1', phrase) #удаляет только один повтор, для удаления других повторов только если еще раз через функцию прогонять
    return(new_phrase)
print(repeat(phrase))
```

+7(xxx)-xxx-xx-xx

Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера. Примеры валидных форматов:

• +7 955 555-55-55;
• 9555555555;
• 8(955)555-55-55;
• +7 955 555 55 55;

• +7 955+555+55+55.

```python
regex_telnumber = r'^((\+?7|8?)[\s\-]?)?((\(\d{3}\))|(\d{3}))?([\-\s\+])?(\d{3})[\-\s\+]?(\d{2})[\-\s\+]?(\d{2})$'
def valid_number(regex_telnumber):
    user_input = input('Введите номер сотового телефона ')
    result = re.search(regex_telnumber, user_input)
    if re.match(regex_telnumber, user_input):
        correct_number = re.sub(regex_telnumber, r'+7(\3)-\7-\8-\9',  user_input) #никак не получилось убрать скобки если они были при вводе, на выходе тогда двойные получаются, в слаке пока не нашла ответ 
        return(correct_number)
    else:
        print('Номер не валиден.')
print(valid_number(regex_telnumber))
```
