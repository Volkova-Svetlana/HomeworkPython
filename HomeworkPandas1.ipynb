```python
import pandas as pd
#page_url = 'https://grouplens.org/datasets/movielens/25m/'
ratings = pd.read_csv('ml-25m/ml-25m/ratings.csv')
movies = pd.read_csv('ml-25m/ml-25m/movies.csv')
joined = ratings.merge(movies, on='movieId', how='left')
filtered_movies = joined[ (joined['rating']==5.0) ]
#filtered_movies.sort_values(by=['movieId'], ascending=[True]).head(50)
filtered_movies.groupby('title').count().sort_values('movieId', ascending=False).head(10)
```

```python
import pandas as pd
data = pd.read_csv('power.csv')
filtered_countries = data[ ((data['country']=='Latvia') | (data['country']=='Lithuania') | (data['country']=='Estonia')) & ((data['category']==12) | (data['category']==4) | (data['category']==12) | (data['category']==21)) & ((data['year']>=2005) & (data['year']<=2010)) & ((data['quantity']>0))]
#filtered_countries.sort_values(by=['quantity', 'country', 'year', 'category'], ascending=[True, True, False, False]).head(50)
summa = filtered_countries[['quantity']].sum()
print(summa)
```
